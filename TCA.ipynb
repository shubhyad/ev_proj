```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("customer churn.csv")
df.head()
```

```python
df.info()
```

#replacing blanks with as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df['TotalCharges'].replace(" ","0")
df["TotalCharges"]=df['TotalCharges'].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen']=df["SeniorCitizen"].apply(conv)
```

#converted 0 to 1 values of senior citizen to yes/no to make it easier to unerstand 

```python
df['customerID'].duplicated().sum()
```

```python
df.head(25)
```

```python
ax=sns.countplot(x=df['Churn'],data=df)
ax.bar_label(ax.containers[0])
plt.title("count of customers churned",fontsize=1)
plt.show()
```

```python
churn_counts = df['Churn'].value_counts()
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.2f%%', startangle=90)
plt.title('Churn Distribution')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```

# from the given pie chart we can conclude that 26.54% of our customers have churned out. #now lets explore the reason behind it

```python
sns.countplot(x='gender',hue="Churn",data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
total_counts = df.groupby( 'SeniorCitizen') ['Churn'].value_counts (normalize=True).unstack() * 100
# Plot
fig, ax = plt.subplots (figsize=(5, 5)) # Adjust figsize for better visualization
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e']) 
# Add percentage Labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y=p.get_xy ( )
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')
plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend (title='Churn', loc='upper right')
plt.show()
```

```python
#comparative a greater pecentage of people in senior citizen category have churned
```

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols = 3
n_rows=(len(columns) + n_cols - 1) // n_cols # Calculate number of rows needed
# Create subplots
fig, axes = plt.subplots (n_rows, n_cols, figsize=(15, n_rows* 4))
axes = axes.flatten()

for i, col in enumerate (columns):
      sns.countplot(x=col, data=df, ax=axes[i],hue=df["Churn"])
      axes[i].set_title(f'Count Plot of {col}')
      axes[i].set_xlabel(col)
      axes[i].set_ylabel('Count')
for j in range(i+1,len(axes)):
      fig.delaxes(axes[j])

plt.tight_layout()
plt.show()      
```

```python
#Customers without security features like OnlineSecurity, OnlineBackup, and TechSupport have higher churn rates.

#Fiber optic internet users show a notably higher churn compared to DSL and no internet service.

#Services like StreamingTV and StreamingMovies are also linked to higher churn when not subscribed to.

#Overall, lack of value-added services correlates with increased churn, suggesting retention could improve by promoting these features.
```
